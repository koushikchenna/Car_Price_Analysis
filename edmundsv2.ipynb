```python
page = requests.get("https://www.edmunds.com/inventory/srp.html?inventorytype=used%2Ccpo&radius=6000&make=ford&model=shelby-gt500", headers={"User-Agent": "Mozilla/5.0 (X11; CrOS x86_64 12871.102.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.141 Safari/537.36"})
print(page)
soup = BeautifulSoup(page.content, 'html.parser')
soup = soup.find('ul', class_='usurp-card-list list-unstyled align-items-stretch row')
soup = soup.find_all('li', class_='d-flex mb-0_75 mb-md-1_5 col-12 col-md-6')
for slide in soup:
    name = slide.find('div', class_='size-16 font-weight-bold text-primary-darker mb-0_5').get_text()
    price = slide.find('span', class_='size-24 font-weight-bold text-gray-darker').get_text()
    link = slide.find('a', class_='usurp-inventory-card-vdp-link')['href']
    print(link)
    print(name)
    print(price)
    names.append(name)
    prices.append(price)
    page2 = requests.get("https://www.edmunds.com" + str(link), headers={"User-Agent": "Mozilla/5.0 (X11; CrOS x86_64 12871.102.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.141 Safari/537.36"})
    soup2 = BeautifulSoup(page2.content, 'html.parser')
    cylinder = soup2.find_all('span', class_='not-opaque text-black')
    print(cylinder[0].get_text())
    cylinders.append(cylinder[0].get_text())
    features = soup2.find_all('div', class_='p-0 col-12 col-md-6')[0]
    LoF = features.find_all('div', class_='m-0 mb-1 row')
    for f in range(0, len(LoF)):
        if "miles" in str(LoF[f].get_text()):  
            mileage.append(LoF[f].get_text())
        if "Ext" in str(LoF[f].get_text()) or "ext" in str(LoF[f].get_text()):  
            exterior_color.append(LoF[f].get_text())
        if "int" in str(LoF[f].get_text()) or "Int" in str(LoF[f].get_text()):  
            interior_color.append(LoF[f].get_text())
        if "transmission" in str(LoF[f].get_text()):  
            transmission.append(LoF[f].get_text())
        print(LoF[f].get_text())
```
